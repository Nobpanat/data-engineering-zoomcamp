```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5433/ny_taxi')
```

```python
df_iter_green_taxi = pd.read_csv(
    'green_tripdata_2019-10.csv.gz',
    dtype = dtype_trip,
    parse_dates=parse_dates_trip,
    iterator=True,
    chunksize=100000
)    
    
```

```python
first = True
from tqdm.auto import tqdm
for df_chunk in tqdm(df_iter_green_taxi):
    if first: 
        df_chunk.head(0).to_sql(
            name="green_taxi",
            con=engine,
            if_exists="replace"
        )
        first = False
        print('Table created')

    df_chunk.to_sql(
            name="green_taxi",
            con=engine,
            if_exists="append"
    )
    print("Inserted:" , len(df_chunk))
```
